```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_univariate_num(df, num_f):
    rows = len(num_f)
    
    plt.figure(figsize=(14, 4 * rows))
    
    for i, col in enumerate(num_f):
        # Boxplot (left
        plt.subplot(rows, 2, 2*i + 1)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        
        # Histogram (right)
        plt.subplot(rows, 2, 2*i + 2)
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
plot_univariate_num(df, num_f)
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
from category_encoders import CatBoostEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression,Lasso
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
high_card = ['EducationField','JobRole']
cat_ft = ['BusinessTravel','Department','Gender','MaritalStatus','OverTime']
```

```python
X = df.drop(columns=['Attrition'])
y = df.iloc[:,1]
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=50)
```

```python
X_train.shape
```

```python
len(num_f)+len(cat_ft)+len(high_card)
```

```python
trf = ColumnTransformer([
    ('sc',StandardScaler(),num_f),
    ('high_cat',CatBoostEncoder(random_state=42),high_card),
    ('cat_ft',OneHotEncoder(sparse_output=False,drop='first'),cat_ft)
],remainder='passthrough')
```

# APPLYING LASSO just for checking the feature Importnace

```python
X_trf = trf.fit_transform(X_train,y_train)
lasso = Lasso(alpha=0.03)
lasso.fit(X_trf,y_train)
feature_names = trf.get_feature_names_out()
coef = pd.Series(lasso.coef_,index=feature_names)
selected_ft = coef[coef != 0].index.to_list()
print(selected_ft)
```

```python
pipe = Pipeline([
    ('trf',trf),
    ('LR',LogisticRegression(max_iter=1500,class_weight='balanced'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
print(f"recall : {np.mean(cross_val_score(pipe, X_train, y_train, scoring='recall'))}")
print(f"PRECISIOON_MACRO := {np.mean(cross_val_score(pipe,X_train,y_train,scoring='precision_macro'))}")
```

```python
import joblib
joblib.dump(pipe,'attrition_pred_model')
```
